# Procurando um banco de dados, segmentando e transformando em dicionário

Os dados foram retirados [Cálculo e Análise de Dieta para o Indivíduo (USP-2018)](https://edisciplinas.usp.br/mod/folder/view.php?id=2301181), mas utilizamos apenas a tab1.xls, afinal, queriamos informações báscias sobre os alimentos, contudo, os alimentos estavam misturados e tivemos que separar manualmente.

Após separar manualmente em diversas tabelas de excel, tivemos que transformá-las em dataframe e por último em diversos dicionários.

```python
import pandas as pd

# Leia o arquivo Excel como um DataFrame
df_acucares = pd.read_excel("Açucares.xlsx")
df_aveseovos = pd.read_excel("Aves e ovos.xlsx")
df_carnes = pd.read_excel("Carnes.xlsx")
df_cereaiseleguminosas = pd.read_excel("Cereais e Leguminosas.xlsx")
df_conservas = pd.read_excel("Conservas.xlsx")
df_farinhasemassas = pd.read_excel("Farinhas e massas.xlsx")
df_frutas = pd.read_excel("Frutas.xlsx")
df_hortalicaselegumes = pd.read_excel("Hortaliças e legumes.xlsx")
df_leitesederivados = pd.read_excel("Leites e derivados.xlsx")
df_paesebolos = pd.read_excel("Pães e bolos.xlsx")

# Exiba os DataFrame
print(df_acucares)
print(df_paesebolos)
```

```python
#Agora precisamos transformar esses df em dicionários(com orientação por linhas);
df_hortalicaselegumes = pd.read_excel("Hortaliças e legumes.xlsx")
df_leitesederivados = pd.read_excel("Leites e derivados.xlsx")
df_paesebolos = pd.read_excel("Pães e bolos.xlsx")

dic_Acucares = df_acucares.to_dict(orient='records')
dic_AvesOvos = df_aveseovos.to_dict(orient='records')
dic_Carnes = df_carnes.to_dict(orient='records')
dic_Cereais_e_Leguminosas = df_cereaiseleguminosas.to_dict(orient='records')
dic_Conservas = df_conservas.to_dict(orient='records')
dic_Farinhas_e_Massas = df_farinhasemassas.to_dict(orient='records')
dic_Frutas = df_frutas.to_dict(orient='records')
dic_Hortalicas_e_Legumes = df_hortalicaselegumes.to_dict(orient='records')
dic_Leite_e_derivados = df_leitesederivados.to_dict(orient='records')
dic_Paes_e_Bolos = df_paesebolos.to_dict(orient='records')
print("Dicionário com orientação por linhas:")
print()
# Exiba os Dicionários
print(dic_Acucares)
print()
print(dic_Paes_e_Bolos)
```
